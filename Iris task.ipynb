```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
irisfile=pd.read_csv("Iris.csv")
```

```python
irisfile.columns
```

```python
irisfile.head()
```

```python
plt.scatter(
    irisfile["SepalLengthCm"],
    irisfile["SepalWidthCm"]
)

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

plt.show()
```

```python
plt.hist2d(
    irisfile["SepalLengthCm"],
    irisfile["SepalWidthCm"],
    bins=30
)

plt.colorbar()

plt.title("Sepal Length vs Sepal Width Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

plt.show()
```

```python
groups = irisfile.groupby("Species")["SepalWidthCm"].apply(list)

plt.boxplot(groups)

plt.xticks([1, 2, 3], irisfile["Species"].unique())

plt.title("Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width")

plt.show()
```
